```python
with open("preguntas.txt", "w", encoding="utf-8") as f:
    f.write("ğŸ§¬ Â¿QuÃ© describe mejor el concepto de polimorfismo en programaciÃ³n orientada a objetos?\n")
    f.write("a) La capacidad de una clase para heredar atributos y mÃ©todos de otra clase.\n")
    f.write("b) La ocultaciÃ³n de los detalles internos de una clase, exponiendo solo lo necesario.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras segÃºn el contexto.\n")
    f.write("d) La creaciÃ³n de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: c\n")
    f.write("\n")
    f.write("ğŸ¶ Â¿CuÃ¡l es la principal ventaja del polimorfismo en la programaciÃ³n orientada a objetos?\n")
    f.write("a) Permite la reutilizaciÃ³n de cÃ³digo mediante la herencia.\n")
    f.write("b) Facilita la creaciÃ³n de interfaces g}rÃ¡ficas de usuario.\n")
    f.write("c) Permite que objetos de diferentes clases respondan al mismo mensaje de manera especÃ­fica.\n")
    f.write("d) Asegura la integridad de los datos mediante la encapsulaciÃ³n.\n")
    f.write("Respuesta correcta: c\n")
    f.write("\n")
    f.write("ğŸ“¦ Â¿QuÃ© tipo de polimorfismo se logra mediante la redefiniciÃ³n de mÃ©todos en una subclase?\n")
    f.write("a) Polimorfismo estÃ¡tico.\n")
    f.write("b) Polimorfismo dinÃ¡mico.\n")
    f.write("c) Sobrecarga de mÃ©todos.\n")
    f.write("d) EncapsulaciÃ³n.\n")
    f.write("Respuesta correcta: b) Polimorfismo dinÃ¡mico.\n")
    f.write("\n")
    f.write("ğŸ” Â¿QuÃ© es una clase abstracta en programaciÃ³n orientada a objetos?\n")
    f.write("a) Una clase que no puede tener mÃ©todos.\n")
    f.write("b) Una clase que no puede ser instanciada directamente.\n")
    f.write("c) Una clase que solo contiene mÃ©todos concretos.\n")
    f.write("d) Una clase que hereda de mÃºltiples clases.\n")
    f.write("Respuesta correcta: b) Una clase que no puede ser instanciada directamente.\n")
    f.write("\n")
    f.write("ğŸ§± Â¿CuÃ¡l es el propÃ³sito de la abstracciÃ³n en la programaciÃ³n orientada a objetos?\n")
    f.write("a) Ocultar los detalles de implementaciÃ³n y mostrar solo la funcionalidad esencial.\n")
    f.write("b) Permitir que una clase herede atributos y mÃ©todos de otra clase.\n")
    f.write("c) Proteger los datos internos de una clase mediante modificadores de acceso.\n")
    f.write("d) Permitir que un objeto tenga mÃºltiples formas o comportamientos.\n")
    f.write("Respuesta correcta: a) Ocultar los detalles de implementaciÃ³n y mostrar solo la funcionalidad esencial.\n")
    f.write("\n")
    f.write("ğŸ”§ Â¿QuÃ© tipo de mÃ©todos puede contener una clase abstracta?\n")
    f.write("a) Solo mÃ©todos concretos.\n")
    f.write("b) Solo mÃ©todos abstractos.\n")
    f.write("c) MÃ©todos abstractos y concretos.\n")
    f.write("d) MÃ©todos estÃ¡ticos Ãºnicamente.\n")
    f.write("Respuesta correcta: c) MÃ©todos abstractos y concretos.\n")
    f.write("\n")
    f.write("ğŸ‘¨â€ğŸ‘©â€ğŸ‘§â€ğŸ‘¦ Â¿QuÃ© es la herencia en programaciÃ³n orientada a objetos?\n")
    f.write("a) La capacidad de una clase para heredar atributos y mÃ©todos de otra clase.\n")
    f.write("b) La ocultaciÃ³n de los detalles internos de una clase, exponiendo solo lo necesario.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras segÃºn el contexto.\n")
    f.write("d) La creaciÃ³n de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: a) La capacidad de una clase para heredar atributos y mÃ©todos de otra clase.\n")
    f.write("\n")
    f.write("ğŸ” Â¿QuÃ© es una subclase en el contexto de la herencia?\n")
    f.write("a) Una clase que hereda atributos y mÃ©todos de otra clase.\n")
    f.write("b) Una clase que no puede ser instanciada directamente.\n")
    f.write("c) Una clase que contiene solo mÃ©todos abstractos.\n")
    f.write("d) Una clase que hereda de mÃºltiples clases.\n")
    f.write("Respuesta correcta: a) Una clase que hereda atributos y mÃ©todos de otra clase.\n")
    f.write("\n")
    f.write("ğŸ”„ Â¿QuÃ© es la herencia mÃºltiple?\n")
    f.write("a) Cuando una clase hereda de una sola clase base.\n")
    f.write("b) Cuando una clase hereda de mÃºltiples clases base.\n")
    f.write("c) Cuando una clase no puede heredar de ninguna clase base.\n")
    f.write("d) Cuando una clase contiene solo mÃ©todos abstractos.\n")
    f.write("Respuesta correcta: b) Cuando una clase hereda de mÃºltiples clases base.\n")
    f.write("\n")
    f.write("ğŸ”’ Â¿QuÃ© es el encapsulamiento en programaciÃ³n orientada a objetos?\n")
    f.write("a) La agrupaciÃ³n de datos con los mÃ©todos que operan en esos datos.\n")
    f.write("b) La capacidad de una clase para heredar atributos y mÃ©todos de otra clase.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras segÃºn el contexto.\n")
    f.write("d) La creaciÃ³n de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: a) La agrupaciÃ³n de datos con los mÃ©todos que operan en esos datos.\n")
    f.write("\n")


from abc import ABC, abstractmethod
import time

# Clase abstracta base
class Pregunta(ABC):
    def __init__(self, texto, opciones, respuesta_correcta):
        self.texto = texto
        self.opciones = opciones
        self.respuesta_correcta = respuesta_correcta

    @abstractmethod
    def mostrar_pregunta(self):
        pass

    @abstractmethod
    def verificar_respuesta(self, respuesta_usuario):
        pass


# Clase que hereda de Pregunta y la implementa
class PreguntaOpcionMultiple(Pregunta):
    def mostrar_pregunta(self):
        print(self.texto)
        for letra, opcion in self.opciones.items():
            print(f"{letra}) {opcion}")

    def verificar_respuesta(self, respuesta_usuario):
        return respuesta_usuario == self.respuesta_correcta


# FunciÃ³n para cargar las preguntas desde el archivo
def cargar_preguntas(archivo):
    preguntas = []
    with open(archivo, "r", encoding="utf-8") as f:
        lines = f.readlines()
        i = 0
        while i < len(lines):
            texto = lines[i].strip()
            opciones = {}
            for j in range(1, 5):
                letra, opcion = lines[i + j].split(") ")
                opciones[letra] = opcion.strip()

            respuesta_correcta = lines[i + 5].split(": ")[1].strip()

            # Crear una instancia de PreguntaOpcionMultiple
            pregunta = PreguntaOpcionMultiple(texto, opciones, respuesta_correcta)
            preguntas.append(pregunta)

            i += 7

    return preguntas


# FunciÃ³n para iniciar el juego
def iniciar_juego():
    print("ğŸ® Bienvenido al juego de preguntas y respuestas ğŸ®")
    nombre = input("ğŸ“ Â¿CuÃ¡l es tu nombre? ")
    print(f"\nğŸ‘‹ Â¡Hola {nombre}! Vamos a comenzar...\n")
    time.sleep(2)

    preguntas = cargar_preguntas("preguntas.txt")
    puntaje = 0

    for pregunta in preguntas:
        pregunta.mostrar_pregunta()
        respuesta_usuario = input("Tu respuesta (a/b/c/d): ").lower()
        if pregunta.verificar_respuesta(respuesta_usuario):
            print("âœ… Â¡Correcto!")
            puntaje += 1
        else:
            print("âŒ Incorrecto.")

    print(f"\nğŸ“Š Puntaje final: {puntaje}/{len(preguntas)}")


# FunciÃ³n principal para ejecutar el juego
if __name__ == "__main__":
    iniciar_juego()
```
