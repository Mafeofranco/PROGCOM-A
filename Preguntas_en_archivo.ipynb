```python
with open("preguntas.txt", "w", encoding="utf-8") as f:
    f.write("🧬 ¿Qué describe mejor el concepto de polimorfismo en programación orientada a objetos?\n")
    f.write("a) La capacidad de una clase para heredar atributos y métodos de otra clase.\n")
    f.write("b) La ocultación de los detalles internos de una clase, exponiendo solo lo necesario.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras según el contexto.\n")
    f.write("d) La creación de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: c\n")
    f.write("\n")
    f.write("🐶 ¿Cuál es la principal ventaja del polimorfismo en la programación orientada a objetos?\n")
    f.write("a) Permite la reutilización de código mediante la herencia.\n")
    f.write("b) Facilita la creación de interfaces g}ráficas de usuario.\n")
    f.write("c) Permite que objetos de diferentes clases respondan al mismo mensaje de manera específica.\n")
    f.write("d) Asegura la integridad de los datos mediante la encapsulación.\n")
    f.write("Respuesta correcta: c\n")
    f.write("\n")
    f.write("📦 ¿Qué tipo de polimorfismo se logra mediante la redefinición de métodos en una subclase?\n")
    f.write("a) Polimorfismo estático.\n")
    f.write("b) Polimorfismo dinámico.\n")
    f.write("c) Sobrecarga de métodos.\n")
    f.write("d) Encapsulación.\n")
    f.write("Respuesta correcta: b) Polimorfismo dinámico.\n")
    f.write("\n")
    f.write("🔍 ¿Qué es una clase abstracta en programación orientada a objetos?\n")
    f.write("a) Una clase que no puede tener métodos.\n")
    f.write("b) Una clase que no puede ser instanciada directamente.\n")
    f.write("c) Una clase que solo contiene métodos concretos.\n")
    f.write("d) Una clase que hereda de múltiples clases.\n")
    f.write("Respuesta correcta: b) Una clase que no puede ser instanciada directamente.\n")
    f.write("\n")
    f.write("🧱 ¿Cuál es el propósito de la abstracción en la programación orientada a objetos?\n")
    f.write("a) Ocultar los detalles de implementación y mostrar solo la funcionalidad esencial.\n")
    f.write("b) Permitir que una clase herede atributos y métodos de otra clase.\n")
    f.write("c) Proteger los datos internos de una clase mediante modificadores de acceso.\n")
    f.write("d) Permitir que un objeto tenga múltiples formas o comportamientos.\n")
    f.write("Respuesta correcta: a) Ocultar los detalles de implementación y mostrar solo la funcionalidad esencial.\n")
    f.write("\n")
    f.write("🔧 ¿Qué tipo de métodos puede contener una clase abstracta?\n")
    f.write("a) Solo métodos concretos.\n")
    f.write("b) Solo métodos abstractos.\n")
    f.write("c) Métodos abstractos y concretos.\n")
    f.write("d) Métodos estáticos únicamente.\n")
    f.write("Respuesta correcta: c) Métodos abstractos y concretos.\n")
    f.write("\n")
    f.write("👨‍👩‍👧‍👦 ¿Qué es la herencia en programación orientada a objetos?\n")
    f.write("a) La capacidad de una clase para heredar atributos y métodos de otra clase.\n")
    f.write("b) La ocultación de los detalles internos de una clase, exponiendo solo lo necesario.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras según el contexto.\n")
    f.write("d) La creación de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: a) La capacidad de una clase para heredar atributos y métodos de otra clase.\n")
    f.write("\n")
    f.write("🔁 ¿Qué es una subclase en el contexto de la herencia?\n")
    f.write("a) Una clase que hereda atributos y métodos de otra clase.\n")
    f.write("b) Una clase que no puede ser instanciada directamente.\n")
    f.write("c) Una clase que contiene solo métodos abstractos.\n")
    f.write("d) Una clase que hereda de múltiples clases.\n")
    f.write("Respuesta correcta: a) Una clase que hereda atributos y métodos de otra clase.\n")
    f.write("\n")
    f.write("🔄 ¿Qué es la herencia múltiple?\n")
    f.write("a) Cuando una clase hereda de una sola clase base.\n")
    f.write("b) Cuando una clase hereda de múltiples clases base.\n")
    f.write("c) Cuando una clase no puede heredar de ninguna clase base.\n")
    f.write("d) Cuando una clase contiene solo métodos abstractos.\n")
    f.write("Respuesta correcta: b) Cuando una clase hereda de múltiples clases base.\n")
    f.write("\n")
    f.write("🔒 ¿Qué es el encapsulamiento en programación orientada a objetos?\n")
    f.write("a) La agrupación de datos con los métodos que operan en esos datos.\n")
    f.write("b) La capacidad de una clase para heredar atributos y métodos de otra clase.\n")
    f.write("c) La capacidad de un objeto de comportarse de diferentes maneras según el contexto.\n")
    f.write("d) La creación de clases que no pueden ser instanciadas directamente.\n")
    f.write("Respuesta correcta: a) La agrupación de datos con los métodos que operan en esos datos.\n")
    f.write("\n")


from abc import ABC, abstractmethod
import time

# Clase abstracta base
class Pregunta(ABC):
    def __init__(self, texto, opciones, respuesta_correcta):
        self.texto = texto
        self.opciones = opciones
        self.respuesta_correcta = respuesta_correcta

    @abstractmethod
    def mostrar_pregunta(self):
        pass

    @abstractmethod
    def verificar_respuesta(self, respuesta_usuario):
        pass


# Clase que hereda de Pregunta y la implementa
class PreguntaOpcionMultiple(Pregunta):
    def mostrar_pregunta(self):
        print(self.texto)
        for letra, opcion in self.opciones.items():
            print(f"{letra}) {opcion}")

    def verificar_respuesta(self, respuesta_usuario):
        return respuesta_usuario == self.respuesta_correcta


# Función para cargar las preguntas desde el archivo
def cargar_preguntas(archivo):
    preguntas = []
    with open(archivo, "r", encoding="utf-8") as f:
        lines = f.readlines()
        i = 0
        while i < len(lines):
            texto = lines[i].strip()
            opciones = {}
            for j in range(1, 5):
                letra, opcion = lines[i + j].split(") ")
                opciones[letra] = opcion.strip()

            respuesta_correcta = lines[i + 5].split(": ")[1].strip()

            # Crear una instancia de PreguntaOpcionMultiple
            pregunta = PreguntaOpcionMultiple(texto, opciones, respuesta_correcta)
            preguntas.append(pregunta)

            i += 7

    return preguntas


# Función para iniciar el juego
def iniciar_juego():
    print("🎮 Bienvenido al juego de preguntas y respuestas 🎮")
    nombre = input("📝 ¿Cuál es tu nombre? ")
    print(f"\n👋 ¡Hola {nombre}! Vamos a comenzar...\n")
    time.sleep(2)

    preguntas = cargar_preguntas("preguntas.txt")
    puntaje = 0

    for pregunta in preguntas:
        pregunta.mostrar_pregunta()
        respuesta_usuario = input("Tu respuesta (a/b/c/d): ").lower()
        if pregunta.verificar_respuesta(respuesta_usuario):
            print("✅ ¡Correcto!")
            puntaje += 1
        else:
            print("❌ Incorrecto.")

    print(f"\n📊 Puntaje final: {puntaje}/{len(preguntas)}")


# Función principal para ejecutar el juego
if __name__ == "__main__":
    iniciar_juego()
```
